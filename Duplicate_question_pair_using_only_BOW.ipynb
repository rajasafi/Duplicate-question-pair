# Here we are going to use different model to see who is giving Best Accuracy

> without doing preprocess



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('/content/train.csv')
```

```python
data.head()
```

```python
new_data=data.sample(30000)
```

```python
new_data.isnull().sum()
```

```python
new_data.duplicated().sum()
```

```python
ques_data=new_data[['question1','question2']]
```

```python
ques_data.head()
```

Now we will convert the data in vector form



> Then we will mergethe question1 and question2 into one list


> After that we will use different model to see who is giving Best Accuracy








```python
from sklearn.feature_extraction.text import CountVectorizer
questions= list(ques_data['question1'])+list(ques_data['question2'])

cv=CountVectorizer(max_features=3000)
q1_arr, q2_arr = np.vsplit(cv.fit_transform(questions).toarray(),2)
```

After splitting the total sampe question into two parts we will now make a dataframe to store the splitted data

```python
temp_dt1=pd.DataFrame(q1_arr, index=ques_data.index)
temp_dt2=pd.DataFrame(q2_arr, index=ques_data.index)
temp_dt=pd.concat([temp_dt1,temp_dt2],axis=1)
temp_dt.shape
```

```python
temp_dt['is_duplicate']= new_data['is_duplicate']
```

```python
temp_dt.head()
```

If we want to see the feature of question1 and question2 after doing BOW

```python
temp_dt
```

#  Here we are going to use different model to see who is giving Best Accuracy

> Random Forest Classifier


> XGB boost Classifier





```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(temp_dt.iloc[:,0:-1].values,temp_dt.iloc[:,-1].values,test_size=0.2,random_state=1)
```

```python
from  sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(X_train,y_train)
y_pred=xgb.predict(X_test)
accuracy_score(y_test,y_pred)
```
